## Ngodarpan

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
import re
```

```python
ngodarpan = pd.read_csv("ngodarpan.csv")
ngodarpan.head()
```

```python
# remove "\n \r \t " in Name & Registration & Address & SectorWorking column
ngodarpan["Name"] = ngodarpan['Name'].replace('\n|\r|\t','', regex=True)
ngodarpan["Registration"] = ngodarpan['Registration'].replace('\n|\r|\t','', regex=True)
ngodarpan["Address"] = ngodarpan['Address'].replace('\n|\r|\t','', regex=True)
ngodarpan["SectorWorking"] = ngodarpan['SectorWorking'].replace('\n|\r|\t','', regex=True)
```

```python
# For the Name and Address Column, make only the first letter of each word be capitalized
ngodarpan["Name"] = ngodarpan["Name"].str.title()
ngodarpan["Address"] = ngodarpan["Address"].str.title()
```

```python
ngodarpan.head()
```

## Registration

### RegNo

```python
# Extract the Registration number from Registration column
ngodarpan['RegNo'] = ngodarpan['Registration'].str.extract(r'(.*),')
ngodarpan.head()
```

### City

```python
# Extract the City from Registration column
ngodarpan['City'] = ngodarpan['Registration'].str.extract(r',(.*)\(')
ngodarpan['City'] = ngodarpan['City'].str.title()
ngodarpan.head()
```

```python
# Visualize the most frequent Cities that appeared (top 5)
city_count = ngodarpan['City'].value_counts().head()
plt.bar(city_count.index, city_count)
```

### State

```python
# Extract the State from Registration column
ngodarpan['State'] = ngodarpan['Registration'].str.extract(r'\((.*)\)')
ngodarpan['State'] = ngodarpan['State'].str.title()
ngodarpan.head()
```

```python
state_count = ngodarpan['State'].value_counts()
state_count
```

```python
# Re-order the columns and Delete the Registration column
ngodarpan = ngodarpan.reindex(['SrNo','Name','RegNo','State','City','Address','SectorWorking'], axis=1) 
ngodarpan.head()
```

## Address

```python
# Extract the PIN from Address column and add this new column after Address
ngodarpan['PIN']=ngodarpan['Address'].str.extract(r'([0-9]{6})')
ngodarpan = ngodarpan.reindex(['SrNo','Name','RegNo','State','City','Address','PIN','SectorWorking'], axis=1) 
```
